### PreProcessing

```python
###changing dtypes of features
```

```python
disney['release_date']=pd.to_datetime(disney['release_date'])
```

```python
disney[['genre','mpaa_rating']]=disney[['genre','mpaa_rating']].astype('category')  ###because both are categoricall values
```

```python
disney.info()
```

```python
disney['Year']=disney['release_date'].dt.year.astype('int64')
```

```python
###how movie genre impacted earnings

```

```python
earnings_genre=disney[['total_gross','inflation_adjusted_gross','genre']].groupby(['genre']).mean().sort_values(by='inflation_adjusted_gross')
display(earnings_genre)
display(earnings_genre.plot(kind='bar'))
```
