```python
cols=ae.select_dtypes(include=[np.object]).columns
```

```python
cols
```

```python
##this methoed is used to convert unreadable format strings into readabel format



ae[cols]=ae[cols].apply(lambda x:x.str.normalize('NFKD').str.encode('ascii',errors='ignore').str.decode('utf-8'))
```

```python
ae.replace({'...':0},inplace=True)
```

```python
ae
```

```python
ae.iloc[0]
```

```python
lst=[]
cols=ae.columns
print(cols)
for i in cols:               ###to convert to numerical
    if i!='State':
        lst.append(i)
print(lst)    
```

```python
ae[lst]=ae[lst].apply(pd.to_numeric)
```

```python
plt.figure(figsize=(30,10))
sns.heatmap(ae.corr(),annot=True,linewidth=1)
```
